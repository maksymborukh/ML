```python
def multiply_m(A, B):
    l_A = len(A)
    l_B = len(B)
    result=[[0 for j in range  (0 , l_A)] for i in range (0 , l_B)]
    for i in range(l_A): 
        for k in range(l_B): 
            r = A[i][k]
            for j in range(len(B[0])): 
                result[i][j] += r * B[k][j]
    return result
```

```python
%timeit multiply_m(matrix_1, matrix_2)
```
